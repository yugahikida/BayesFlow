```python
import pickle
import matplotlib.pyplot as plt
import os
```

```python
PATH = "using_encoder_test"
```

```python
with open(os.path.join('results', PATH, 'tau.pkl'), 'rb') as file:
    tau = pickle.load(file)
```

```python
with open(os.path.join('results', PATH, 'design_loss.pkl'), 'rb') as file:
    design_loss = pickle.load(file)
```

```python
with open(os.path.join('results', PATH, 'bf_loss.pkl'), 'rb') as file:
    bf_loss = pickle.load(file)["loss"]
```

```python
N_steps_1 = len(bf_loss)
N_steps_2 = 500
N_steps_3 = 500
```

### Loss curve for step_1

```python
len(bf_loss)
```

```python
plt.plot(range(N_steps_1), bf_loss)
```

### Loss curve for step_2

```python
step2_loss = design_loss[:N_steps_2][0::10] # every 10th element
step2_loss = design_loss[:N_steps_2]

plt.plot(range(len(step2_loss)), step2_loss)
plt.show()
```

```python
# by specific tau
T = 20

for t in range(T):
    loss_tau = [design_loss[i] for i in range(N_steps_2) if tau[i] == t]
    plt.plot(range(len(loss_tau)), loss_tau, label=f"tau={t}")
plt.legend()
plt.show()
```

### Loss curve for step_3

```python
plt.plot(range(N_steps_3), design_loss[N_steps_2:N_steps_2 + N_steps_3])
```
